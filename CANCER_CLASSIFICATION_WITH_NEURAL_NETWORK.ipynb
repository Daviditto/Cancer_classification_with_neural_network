# build and trian the model 

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X= df.drop('benign_0__mal_1', axis=1).values
y = df['benign_0__mal_1'].values
```

```python
Xtrain, Xtest, ytrain ,ytest= train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(Xtrain)
```

```python
X.shape
```

```python
Xtrain= scaler.transform(Xtrain)
Xtest= scaler.transform(Xtest)
```

```python
network_1 = Sequential()
network_1.add(Dense(30, activation='relu'))
network_1.add(Dense(15, activation='relu'))
network_1.add(Dense(1, activation='sigmoid'))
network_1.compile(optimizer='adam', 
                 loss='binary_crossentropy')
```

```python
network_1.fit(Xtrain, ytrain, validation_data=(Xtest, ytest), epochs=600,verbose=3)
```

```python
history = pd.DataFrame(network_1.history.history)
```

```python
history.plot(); # this is a classic example of overfitting, since the val_loss is increasing and the loss is still dropping
```

```python
# let's do the early stopping to prevent overfitting 
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
network = Sequential()
network.add(Dense(30, activation='relu'))
network.add(Dense(15, activation='relu'))
network.add(Dense(1, activation='sigmoid'))
network.compile(optimizer='adam', 
                 loss='binary_crossentropy')
```

```python
earlystop = EarlyStopping(monitor='val_loss', mode='min', patience=25, verbose=1)
```

```python
network.fit(Xtrain, ytrain, validation_data=(Xtest, ytest), epochs=600,verbose=3, callbacks=[earlystop])
```

```python
history_2 = pd.DataFrame(network.history.history)
```

```python
history_2.plot(); # this is perfect !!!
```

```python
# additionally, we can also do dropout layers to prevent overfitting 
```

```python
from tensorflow.keras.layers import Dropout
```

```python
network_dropout = Sequential()
network_dropout.add(Dense(30, activation='relu'))
network_dropout.add(Dropout(0.5))
network_dropout.add(Dense(15, activation='relu'))
network_dropout.add(Dropout(0.5))


network_dropout.add(Dense(1, activation='sigmoid'))
network_dropout.compile(optimizer='adam', 
                 loss='binary_crossentropy')
```

```python
network_dropout.fit(Xtrain, ytrain, validation_data=(Xtest, ytest), epochs=600,verbose=3, callbacks=[earlystop])
```

```python
history_3 = pd.DataFrame(network_dropout.history.history)
```

```python
history_3.plot(); #this is also really good !!!
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
predictions = network_dropout.predict_classes(Xtest)
```

```python
confusion_matrix(ytest, predictions)
```

```python
print(classification_report(ytest, predictions))
```
